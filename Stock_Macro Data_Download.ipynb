```python

# Get to Diectory import files in a loop 
import os 
#To Cycle through csv files
import glob 

#Get and print working dir
os.getcwd()

# All files in dir
os.listdir()

# Folders and current notebook in sale folder
folder = "." 

stock_macro_array_df = {}
#For each file ends in csv, go through cyclet
for file in glob.glob("*.csv"):
    # Skip s&p because has more date 
    if file == 'sp500.csv': 
        print("Skipping file S&P")
        continue
        
    #Save file name and replace csv with space for col name
    filename = file
    colname = filename.replace(".csv", "")

    # Read current CSV
    curr_df = pd.read_csv(file)

    #Change first col = date and second col = name of file
    curr_df.columns = ['date', colname]

    # Store new individual df in dictionary
    stock_macro_array_df[colname] = curr_df
    #print(stock_macro_df)
```

```python

# Join DF array to Merge all DF  into 1 on data

# Save each data frame values to allow a merge
each_df_values = list(stock_macro_array_df.values())

# Start with first df
stock_macro_df = each_df_values[0]

# Update the current df, merge with next
for next_df in each_df_values[1:]:
    stock_macro_df = pd.merge(stock_macro_df, next_df, on = 'date', how = 'outer')

print(stock_macro_df)
stock_macro_df.head(20)
```

```python
# See a lot of null, date is an object 
stock_macro_df.info()
```

```python
# Read in SP500 Data annd add to data frame
sp500 = yf.download("^GSPC", start="2006-01-01", end="2010-12-21", auto_adjust=True)


#Retrieve only SP data, drop ticket level
sp500 = sp500.xs("^GSPC", axis=1, level=1)

#Reset index return new df
sp500 = sp500.reset_index()
```

```python
#Print Information
sp500.info()
sp500.head()

# Copy over and keep just data and close columns
sp500 = sp500[['Date', 'Close']]  
sp500.columns = ['date', 'sp500']
```

```python
### sp500 left join filters to be only dates that listen in spt
#Filter by date January 2006 to December 2010
#from datetime import datetime, timedelta

#Change date object -> datetime dtype
stock_macro_df['date'] = pd.to_datetime(stock_macro_df['date'])
```

```python
# Join stock data and macro df on data 
# Left join, sp500 filtered from jan 2006 to dec 2010
merged_df = pd.merge(sp500, stock_macro_df, on = 'date', how = 'outer')

merged_df.head()
```

```python

#Date borders (need first of the month data)
start_date = pd.to_datetime('2006-01-01')
end_date = pd.to_datetime('2010-12-31')
#Filter by date
filtered_stock_macro_df = merged_df[merged_df['date'].between(start_date, end_date)]

#print(filtered_stock_macro_df.head(20))

#filtered_stock_macro_df.to_csv('filtered_stock_macro_df.csv', index = False)
```

```python
filtered_stock_macro_df.head(20)

# Order by date 
sorted_df = filtered_stock_macro_df.sort_values(by="date")


#Each of the indecies and rates are determined monthly/quarterly 
#Use most recent valid date to fill in ones after with forward fit 
filled_df = sorted_df.ffill()

#Forward fill on all macroeconomic and on weekends do not want gaps 
filled_df.to_csv('filled_df.csv', index = False)
filled_df.info()
```

```python
# Drop 2 NA rows in sp500
filled_df = filled_df.dropna(subset=['sp500'])
filled_df.info()
```

```python
filled_df.head(20)
#filled_df.to_csv('filled_df.csv', index = False)

#Rename DF 
stock_macrotrends_df = filled_df
```

```python
#Compute 200 day rolling average 
stock_macrotrends_df["move_avg_200day"] = stock_macrotrends_df['sp500'].rolling(window=200).mean()

# If todays price above moving avergae, have a buy signal(1), else sell signal(0)
stock_macrotrends_df["signal_today"] = np.where(stock_macrotrends_df["sp500"] > stock_macrotrends_df["move_avg_200day"], 1, 0)

stock_macrotrends_df.head(10)
```

```python
# Delete the first 200 days, got neded data and signals, (Todays Signals)
# Now start in July 21st 2006
stock_macro_final_df = stock_macrotrends_df.dropna(subset=['move_avg_200day'])
stock_macro_final_df.head(50) 
```

```python
# Need to predict tomorrows signal using todays information! (cant use todays close price to predict buy or sell
stock_macro_final_df['target_signal'] = stock_macro_final_df['signal_today'].shift(-1)

#Drop the last day NA
stock_macro_shifted_df = stock_macro_final_df.dropna(subset=['target_signal'])

stock_macro_shifted_df.head(5) 
```

```python
# Export and print final CSV file
stock_macro_shifted_df.to_csv('stock_macro_shifted_df.csv', index = False)
```

```python
######################################
#####  LOGISTIC REGRESSION   #########
######################################
#Importing Libraries

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab
import scipy.stats as st

#Importing Logistic Regression 
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
stock_macro_shifted_df.columns
```

```python
# Creating independednt(predictor) and dependtent variables (target)
```

```python
# predictors = macroeconomic factors, will not include data sp500 price, or todays signals
X_macro_features = stock_macro_shifted_df[['crude_oil_price', 'real_estate_index', 'retail_sales',
       'federal_interest_rate', 'gold_price', 'unemployment_rate',
       'inflation_rate', 'tech_investment', 'gdp', 'corporate_profits']]

today_signal_targ = stock_macro_shifted_df['signal_today']
y_signal_targ = stock_macro_shifted_df['target_signal']


```

```python
# Splitting data into training and testing 70% train, 30% test
X_macro_train, X_macro_test, y_targ_train, y_targ_test = train_test_split(X_macro_features, y_signal_targ, test_size = 0.3, random_state = 42)
```

```python
# Initialize Logistic Regression object with scaling
logreg_model = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
```

```python
# Fitt the model n trainign data using fit method  (need pipieline to scale at same time_
logreg_model.fit(X_macro_train, y_targ_train)
```

```python
#Make predictions 
y_signal_pred = logreg_model.predict(X_macro_test)
print(y_signal_pred)
```

```python
#Evaluating the Model

#Creating Confusion Matrix 

cm = confusion_matrix(y_targ_test,y_signal_pred)
conf_matrix = pd.DataFrame(data = cm, columns = ['Predicted:0', 'Predicted:1'], index = ['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap="YlGnBu")
plt.show()
                          
```

```python
TN=cm[0,0]
TP=cm[1,1]
FN=cm[1,0]
FP=cm[0,1]
```

```python
#Accuracy
print('The acuracy of the model = TP+TN/(TP+TN+FP+FN) = ', round((TP+TN)/float(TP+TN+FP+FN),3))
```

```python
#Misclassification
print('The Missclassification = 1-Accuracy = ', round(1-((TP+TN)/float(TP+TN+FP+FN)), 3))
```

```python
#Sensitivity
print('Sensitivity or True Positive Rate = TP/(TP+FN) = ', round(TP/float(TP+FN),3))
```

```python
#Specificity
print('Specificity or True Negative Rate = TN/(TN+FP) = ', round(TN/float(TN+FP),3))
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_targ_test,y_signal_pred))
```

```python
#AUC-ROC Curve 
from sklearn.metrics import roc_curve, auc
y_pred_proba = logreg_model.predict_proba(X_macro_test)

# compute roc curve values
fpr, tpr, thresholds = roc_curve(y_targ_test, y_pred_proba[:,1])

# Compute Accuracy 
roc_auc = auc(fpr, tpr)

# plot fpr and tpr axises 
plt.plot(fpr, tpr, label=f"AUC={roc_auc:.3f}", color = '#2ecc71', linewidth = 1.5)

# If was Random choice diagonal line
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

plt.title('ROC Curve for Stock Buy Signal Classifier')
plt.xlabel('False positive rate (1-Specificity)')
plt.ylabel('True positive rate (Sensitivity)')
plt.legend(loc ='lower right')
plt.grid(True)
```
